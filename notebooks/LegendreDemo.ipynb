```python
%matplotlib inline
import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt
```

```python
# Define domain, [-1, 1]
x = np.linspace(-1.0,1.0,1000)

# Make some Legendre polynomials...
P0 = np.polynomial.legendre.Legendre((1,))
P1 = np.polynomial.legendre.Legendre((0,1))
P2 = np.polynomial.legendre.Legendre((0,0,1))
P3 = np.polynomial.legendre.Legendre((0,0,0,1))
P4 = np.polynomial.legendre.Legendre((0,0,0,0,1))
P5 = np.polynomial.legendre.Legendre((0,0,0,0,0,1))
P6 = np.polynomial.legendre.Legendre((0,0,0,0,0,0,1))

# Plot the polynomials
plt.plot(x,P0(x), x,P1(x), x,P2(x), x,P3(x), x,P4(x))
plt.xlabel("x")
plt.ylabel("Pn(x)")
```

```python
# Let's approximate the following function:
f = lambda x : np.sin(x*np.pi)

# First, compute the moments/coefficients
f0 = integrate.quad(lambda x: f(x)*P0(x), -1,1)[0]
f1 = integrate.quad(lambda x: f(x)*P1(x), -1,1)[0]
f2 = integrate.quad(lambda x: f(x)*P2(x), -1,1)[0]
f3 = integrate.quad(lambda x: f(x)*P3(x), -1,1)[0]
f4 = integrate.quad(lambda x: f(x)*P4(x), -1,1)[0]
f5 = integrate.quad(lambda x: f(x)*P5(x), -1,1)[0]
f6 = integrate.quad(lambda x: f(x)*P6(x), -1,1)[0]
fc = [f0,f1,f2,f3,f4,f5,f6]
```

```python
print fc
```

```python
# Now do the expansion
fLegApprx = 0.0
P = [P0, P1, P2, P3, P4, P5, P6]
for n in range(6) :
    fLegApprx += (2.0*n + 1.0)/2.0*fc[n]*P[n](x)
```

```python
plt.plot(x,fLegApprx, x,f(x))
```
